Carga imagen y convierte a RGB

```python
#Carga imagen ejemplo con monedas
img = cv2.imread('Monedas.jpg') 
print(img.shape)
#Recordar que OpenCV lee las im√°genes en BGR, por lo que convertimos para visualizr a RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.axis("off")
plt.imshow(img_rgb) 
plt.show()
```

Convierte a gris y muestra histograma

```python
#Convierte la imagen a todos de gris, mostrando el resultado
img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#C√°lculo del histograma con 256 bins de una imagen en escala de grises
hist = cv2.calcHist([img_gris], [0], None, [256], [0, 256])

plt.figure()
plt.subplot(1, 2, 1)
plt.axis("off")
plt.imshow(img_gris, cmap='gray')

# Histograma sin normalizar
plt.subplot(1, 2, 2)
plt.title("Histograma")
plt.xlabel("Bins")
plt.ylabel("# p√≠xeles")
plt.plot(hist)
plt.xlim([0, 256])
# Separo subplots horizontalmente
plt.subplots_adjust(wspace=0.4)
```

Cuenta resultado tras umbralizar (poniendo a negro el fondo)

```python
#Dos variantes de umbralizado. Probar otros par√°metros, aplicar filtro previo, etc.
umbral = 200 # Prueba varios comenzando en 130
# Umbralizado binario invertido, dado que por defecto se asume objetos en blanco
th1,img_th1 = cv2.threshold(img_gris,umbral,255,cv2.THRESH_BINARY_INV)
print('Umbral fijo usado ', th1)
# Umbralizado con m√©todo de Otsu para selecci√≥n autom√°tica del umbral
th2,img_th2 = cv2.threshold(img_gris,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
print('Umbral Otsu ', th2)

plt.subplot(1, 3, 1)
plt.axis("off")
plt.imshow(img_gris,cmap='gray') 
plt.title('Original')

plt.subplot(1, 3, 2)
plt.axis("off")
plt.imshow(img_th1,cmap='gray') 
plt.title('FIJO invertida')

plt.subplot(1, 3, 3)
plt.axis("off")
plt.imshow(img_th2,cmap='gray') 
plt.title('OTSU invertida')

```

Alternativa contando c√≠rculos utilizando la Transformada de Hough. La selecci√≥n de par√°metros puede ser "divertida", m√°s [informaci√≥n](https://docs.opencv.org/4.x/da/d53/tutorial_py_houghcircles.html)


```python
#Conversi√≥n a gris
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#Suaviza imagen (elimina altas frecuencias)
pimg = cv2.medianBlur(gris, 7)
#Localiza C√≠rculos
circ = cv2.HoughCircles(
        pimg,  # imagen 
        cv2.HOUGH_GRADIENT,  # tipo de detecci√≥n
        1,
        100,  # distancia m√≠nima entre c√≠rculos
        param1=100, # valor del gradiente
        param2=50, # umbral acumulador
        minRadius=50,  # radio m√≠nimo
        maxRadius=150,  # radio m√°ximo
    )

#Dibuja sobre entrada e imagen vac√≠a
img_cont = np.zeros(img_rgb.shape)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
for det in circ[0]:
        x_coor, y_coor, det_radio = det
        cv2.circle(img_rgb,(int(x_coor), int(y_coor)),
            int(det_radio),(0, 255, 0), 2)
        cv2.circle(img_cont,(int(x_coor), int(y_coor)),
            int(det_radio),(255, 255, 255), -1)

#Muestra resultado
plt.subplot(121)
plt.axis("off")
plt.imshow(img_rgb) 
plt.title('C√≠rculos')

plt.subplot(122)
plt.axis("off")
plt.imshow(img_cont) 
plt.title('Rellenos')

plt.show()
```

Las formas localizadas tienen distintas caracter√≠sticas geom√©tricas, que pueden estimarse a partir de sus contornos. M√°s informaci√≥n en la [documentaci√≥n de OpenCV](https://docs.opencv.org/4.x/dd/d49/tutorial_py_contour_features.html).

```python
#Creaci√≥n de pol√≠gono regular
def poligono_regular(image, ctr, radio, lados, color):
    pts = []
    ang_step = 2 * np.pi / lados
    for i in range(lados):
        ang = i * ang_step
        x = int(ctr[0] + radio * np.cos(ang))
        y = int(ctr[1] + radio * np.sin(ang))
        pts.append((x, y))
    pts = np.array(pts, np.int32)
    #regorganiza
    pts = pts.reshape((-1, 1, 2))
    cv2.fillPoly(image, [pts], color=color)

# Imagen vac√≠a
img = np.zeros((400, 400, 1), dtype="uint8")
color = (255, 255, 255)

# Formas b√°sicas
cv2.circle(img, (100, 100), 50, color, -1)  # Circular
poligono_regular(img, (250, 150), 50, 5, color)  # Pol√≠gono regular
cv2.ellipse(img, (200, 300), (100, 40), 0, 0, 360, color, -1)  # El√≠ptica

# Localiza contornos
contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Par√°meros texto
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.3
thickness = 1

# Process each contour to calculate compactness and ellipse ratio (if possible)
for c in contours:
    # Puntos del contorno
    clon = len(c)

    # √Årea y per√≠metro
    area = cv2.contourArea(c)
    perimetro = cv2.arcLength(c, True)

    #Contenedor alineado con ejes de la imagen
    x,y,w,h = cv2.boundingRect(c)
    
    # Compactness: 4*pi*Area/Perimeter^2
    if perimetro > 0:
        compacidad = (perimetro ** 2) / area
    else:
        compactness = 0
    
    # Ajusta elipse si hay suficientes puntos
    if clon >= 5:
        elipse = cv2.fitEllipse(c)
        (center, axes, orientation) = elipse
        major_axis = max(axes)
        minor_axis = min(axes)
        elipse_ratio = major_axis / minor_axis
    else:
        elipse_ratio = None
    
    # Muestra valores en imageb
    cv2.putText(img, f"n: {clon:.1f}", (x, int(y+h+10)), font, font_scale, (255, 255, 255), thickness)
    cv2.putText(img, f"A: {area:.1f}", (x, int(y+h+20)), font, font_scale, (255, 255, 255), thickness)
    cv2.putText(img, f"P: {perimetro:.1f}", (x, int(y+h+30)), font, font_scale, color, thickness)
    cv2.putText(img, f"C: {compacidad:.1f}", (x, int(y+h+40)), font, font_scale, color, thickness)
    cv2.putText(img, f"ER: {elipse_ratio:.1f}", (x, int(y+h+50)), font, font_scale, color, thickness)
    
# Visualiza la imagen
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Formas")
plt.axis('off')
plt.show()
```

Nota: Para establecer la correspondencia entre p√≠xeles y mil√≠metros, comentar que la moneda de un euro tiene un di√°metro de 23.25 mm. la de 50 c√©ntimos de 24.35, la de 20 c√©ntimos de 22.25, etc.

```python
# Clasificaci√≥n simple por color medio: cobre vs amarillo
# Definimos funci√≥n para clasificar seg√∫n los valores RGB

def clasificar_color(mean_color):
    R, G, B = mean_color[2], mean_color[1], mean_color[0]
    # Cobre: R claramente mayor que G, y B bajo
    # Amarillo: R y G altos y similares, B bajo
    if ((G - B > 30) and (G - B <50)) and ((R-G > 70) and (R - G <90)):
        return "Cobre"
    elif ((G-B > 60) and (G - B < 90)) and ((R - G > 30) and (R - G < 50)):
        return "Amarillo"
    else:
        return "Otro"

# Radios reales de monedas (mm)
real_radii = {
    1: 16.25, # 1 c√©ntimo
    2: 18.75,   # 2 c√©ntimos
    5: 21.25,   # 5 c√©ntimos
    10: 19.75,  # 10 c√©ntimos
    20: 22.25,  # 20 c√©ntimos
    50: 24.25,  # 50 c√©ntimos
    100: 23.25, # 1 euro
    200: 25.75, # 2 euros
}

# Funci√≥n para manejar clics y seleccionar una moneda como 1 euro
def click_event(event, x, y, flags, param):
    global moneda_1euro
    if event == cv2.EVENT_LBUTTONDOWN:
        if circs is not None:
            for idx, i in enumerate(circs[0, :]):
                cx, cy, r = i
                dist = math.hypot(x - cx, y - cy)
                if dist <= r:
                    moneda_1euro = (cx, cy, r)
                    cv2.circle(img_gris_cont, (cx, cy), r, (255, 0, 0), 3)
                    cv2.putText(img_gris_cont, "1 Euro", (cx - 40, cy - r - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
                    cv2.imshow("Selecciona una moneda", img_gris_cont)
                    cv2.waitKey(500)
                    cv2.destroyAllWindows()
                    return
```

```python
# --- Cargar imagen ---
img = cv2.imread('Monedas.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
pimg = cv2.medianBlur(img_gris, 5)
valor_total = 0.0

# --- Detecci√≥n de c√≠rculos ---
circs = cv2.HoughCircles(
    pimg, cv2.HOUGH_GRADIENT, 1, 100, param1=100, param2=50,
    minRadius=50, maxRadius=150
)
img_gris_cont = cv2.cvtColor(pimg, cv2.COLOR_GRAY2BGR)
circles_radii = {}
if circs is not None:
    circs = np.uint16(np.around(circs))
    for idx, i in enumerate(circs[0, :], start=1):
        x, y, r = i
        circles_radii[idx] = r

# --- Selecci√≥n de la moneda de 1 ‚Ç¨ ---
moneda_1euro = None
cv2.namedWindow("Selecciona una moneda", cv2.WINDOW_NORMAL)
cv2.imshow("Selecciona una moneda", img)
cv2.setMouseCallback("Selecciona una moneda", click_event)
cv2.waitKey(0)
cv2.destroyAllWindows()

# --- Clasificaci√≥n por tama√±o y color ---
if moneda_1euro is not None:
    clasificaciones_finales = []
    for idx, r in circles_radii.items():
        cx, cy, _ = circs[0, idx - 1]
        radio_mm = r / moneda_1euro[2] * real_radii[100]  # Escalado a mm
        mask = np.zeros(img_gris.shape, dtype="uint8")
        cv2.circle(mask, (int(cx), int(cy)), int(r), 255, -1)
        mean_color = cv2.mean(img, mask=mask)[:3]  # (B, G, R)
        color = clasificar_color(mean_color)

        # Diferencias con radios reales
        diferencias = [(valor, abs(radio_mm - real_r)) for valor, real_r in real_radii.items()]
        diferencias.sort(key=lambda x: x[1])
        # Tomar las dos opciones m√°s cercanas
        top2 = diferencias[:2]
        # Por defecto, la m√°s cercana
        valor_mas_probable = top2[0][0]

        # Usar color solo como ayuda entre las dos m√°s cercanas
        if color == "Cobre":
            opciones_cobre = [v for v, _ in top2 if v in [1, 2, 5]]
            if opciones_cobre:
                valor_mas_probable = opciones_cobre[0]
        elif color == "Amarillo":
            opciones_amarillo = [v for v, _ in top2 if v in [10, 20, 50]]
            if opciones_amarillo:
                valor_mas_probable = opciones_amarillo[0]
        elif color == "Gris":
            opciones_gris = [v for v, _ in top2 if v in [100, 200]]
            if opciones_gris:
                valor_mas_probable = opciones_gris[0]
        # Dibujar resultado
        cv2.circle(img_rgb, (int(cx), int(cy)), int(r), (0, 255, 0), 2)
        texto = f"{valor_mas_probable/100:.2f} euros ({color})"
        valor_total += valor_mas_probable / 100.0
        cv2.putText(img_rgb, texto, (int(cx)-80, int(cy)+int(r)+20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
    # --- Mostrar resultado final ---
    plt.axis("off")
    plt.imshow(img_rgb)
    plt.title(f"Valor total de las monedas: {valor_total:.2f} euros")
    plt.show()

else:
    print("No se seleccion√≥ la moneda de 1 euro.")
```

El objetivo de la siguiente tarea, descrita m√°s abajo, es desarrollar tu propio clasificador basado √∫nicamente en heur√≠sticas basadas en caracter√≠sticas geom√©tricas y/o de apariencia, a partir de las im√°genes completas de las part√≠culas de cada tipo, debiendo mostrar la bondad del clasificador haciendo uso de m√©tricas para ello sobre las anotaciones de la imagen de test *MPs_test.png*. La siguiente celda obtiene varias m√©tricas para un conjunto de datos imaginario (y con etiquetas aleatorias). Si bien las trataremos con m√°s detalle en teor√≠a, muestro un repertorio de ellas, dando m√°s peso a la matriz de confusi√≥n. La ejecuci√≥n de la celda requiere instalar el paquete scikit-learn.

¬øQu√© es una matriz de confusi√≥n?
Se utiliza para mostrar el comportamiento de un clasificador para las distintas clases conocidas, se relacionan las etiquetas de las muestras anotadas frente a las predichas por el clasificador. Se busca una matriz diagonal, pero la perfecci√≥n es infrecuente.

El siguiente ejemplo, muestra el modo de obtener la matriz de confusi√≥n para un hipot√©tico problema con cuatro clases, y valores de anotaci√≥n (variable y) y predicci√≥n (variable y_pred) obtenidos de forma aleatoria.

```python

import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)


# Numero de muestras
n = 100  
nclases = 4

# A falta de clasificador y conjunto de datos, creamos anotaciones y predicciones de forma aleatoria
# Vector aleatorio con etiquetas anotadas
y = [random.randint(0, nclases - 1) for _ in range(n)]
print('Anotaciones ' , y)

# Vector aleatorio con etiquetas predichas por un supuesto clasificador
y_pred = [random.randint(0, nclases - 1) for _ in range(n)]
print('Predicciones ' , y_pred)

print('¬øC√≥mo de bien encajan anotaci√≥n y predicci√≥n?')

#C√°lculo de m√©tricas
accuracy = accuracy_score(y, y_pred)
#Para m√°s de una clase se define la forma de promediar
precision = precision_score(y, y_pred,average='weighted')
recall = recall_score(y, y_pred,average='weighted')
f1score = f1_score(y, y_pred,average='weighted')

print(f"Accuracy (TP/(n))= {accuracy}")
print(f"Precision (TP/(TP+FP)) = {precision}")
print(f"Recall (TP/(TP+FN)) = {recall}")
print(f"F1 Score (2*(precision*recall)/(precision+recall)) = {f1score}")


conf_matrix = confusion_matrix(y, y_pred)
plt.figure(figsize=(8,8))
sns.set(font_scale = 1.75)#tama√±os tipograf√≠a
sns.set(font_scale = 3.0)

ax = sns.heatmap(
        conf_matrix, # confusion matrix 2D array 
        annot=True, # Muestra n√∫meros en las celdas
        fmt='d', # valores enteros
        cbar=False, # sin barra de colores
        cmap='flag', # mapa de colores
        #vmax=175 # contraste de color
    )

#Etiquetas matriz de confusi√≥n
label_font = {'size':'25'}
ax.set_xlabel("Predicha", labelpad=-0.75, fontdict=label_font)
ax.set_ylabel("Real/Anotado", labelpad=20, fontdict=label_font)
```

TAREA: La tarea consiste en extraer caracter√≠sticas (geom√©tricas y/o visuales) de las tres im√°genes completas de partida, y *aprender* patrones que permitan identificar las part√≠culas en nuevas im√°genes. Para ello se proporciona como imagen de test *MPs_test.jpg* y sus correpondientes anotaciones *MPs_test_bbs.csv* con la que deben obtener las m√©tricas para su propuesta de clasificaci√≥n de micropl√°sticos, adem√°s de la matriz de confusi√≥n. La matriz de confusi√≥n permitir√° mostrar para cada clase el n√∫mero de muestras que se clasifican correctamente de dicha clase, y el n√∫mero de muestras que se clasifican incorrectamente como perteneciente a una de las otras dos clases.

En el trabajo [SMACC: A System for Microplastics Automatic Counting and Classification](https://doi.org/10.1109/ACCESS.2020.2970498), las caracter√≠sticas geom√©tricas utilizadas fueron:

- √Årea en p√≠xeles
- Per√≠metro en p√≠xeles
- Compacidad (relaci√≥n entre el cuadrado del per√≠metro y el √°rea de la part√≠cula)
- Relaci√≥n del √°rea de la part√≠cula con la del contenedor
- Relaci√≥n del ancho y el alto del contenedor
- Relaci√≥n entre los ejes de la elipse ajustada
- Definido el centroide, relaci√≥n entre las distancias menor y mayor al contorno

Si no se quedan satisfechos con la segmentaci√≥n obtenida, es el mundo real, tambi√©n en el README comento t√©cnicas recientes de segmentaci√≥n, que podr√≠an despertar su curiosidad.

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import cv2
import joblib

# -----------------------------
# Argumentos simulando tu clase Args
# -----------------------------
class Args:
    def __init__(self):
        self.train_images = ['FRA.png', 'PEL.png', 'TAR.png']  # Ajusta seg√∫n tus archivos
        self.train_csvs = ['MPs_test_bbs.csv']                 # CSV con todas las anotaciones
        self.output_dir = 'output_model'
        self.debug = True

# -----------------------------
# Funciones de soporte
# -----------------------------
def extract_features_from_image(image_path, gt_df):
    """
    Extrae features de cada bounding box en la imagen seg√∫n el CSV de GT.
    Retorna X (features) y y (labels).
    """
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if img is None:
        raise ValueError(f"No se pudo leer la imagen {image_path}")

    # Extraer etiquetas del CSV para esta imagen
    label_name = os.path.splitext(os.path.basename(image_path))[0]
    gt_for_image = gt_df[gt_df['label'] == label_name]

    X = []
    y = []

    for _, row in gt_for_image.iterrows():
        x_min, y_min, x_max, y_max = int(row['x_min']), int(row['y_min']), int(row['x_max']), int(row['y_max'])
        roi = img[y_min:y_max, x_min:x_max]

        # -----------------------------
        # Extracci√≥n de features simple
        # -----------------------------
        # Ejemplo: color promedio en RGB
        mean_colors = cv2.mean(roi)[:3]  # (B, G, R)
        # Agregar otras features si quieres, aqu√≠ simple 3 valores
        X.append(mean_colors)
        y.append(label_name)

    return np.array(X), np.array(y)

# -----------------------------
# Funci√≥n principal
# -----------------------------
def main(args):
    os.makedirs(args.output_dir, exist_ok=True)

    # Leer CSV √∫nico
    gt_df = pd.read_csv(args.train_csvs[0])

    X_train_list = []
    y_train_list = []

    for img_path in args.train_images:
        if args.debug:
            print(f"Procesando imagen: {img_path}")

        X_img, y_img = extract_features_from_image(img_path, gt_df)

        if len(X_img) == 0:
            print(f" No hay anotaciones para {img_path}, se salta")
            continue

        X_train_list.append(X_img)
        y_train_list.append(y_img)

    # Concatenar todo
    if len(X_train_list) == 0:
        raise ValueError("No se han extra√≠do features de ninguna imagen.")

    X_train = np.vstack(X_train_list)
    y_train = np.concatenate(y_train_list)

    if args.debug:
        print(f"Features de entrenamiento: {X_train.shape}, Labels: {y_train.shape}")

    # -----------------------------
    # Entrenar clasificador
    # -----------------------------
    clf = Pipeline([
        ('scaler', StandardScaler()),
        ('rf', RandomForestClassifier(n_estimators=200, random_state=42))
    ])

    print("Entrenando clasificador...")
    clf.fit(X_train, y_train)

    # Guardar modelo entrenado
    joblib.dump(clf, os.path.join(args.output_dir, 'trained_model.joblib'))
    print(f"Modelo entrenado y guardado en {args.output_dir}/trained_model.joblib")

    # -----------------------------
    # Evaluaci√≥n con matriz de confusi√≥n
    # -----------------------------
    y_pred = clf.predict(X_train)
    labels_order = ['PEL', 'FRA', 'TAR']
    cm = confusion_matrix(y_train, y_pred, labels=labels_order)

    print("Matriz de confusi√≥n:")
    print(cm)
    print(classification_report(y_train, y_pred, labels=labels_order))

    # Visualizaci√≥n con n√∫meros
    plt.figure(figsize=(5,5))
    plt.imshow(cm, cmap='Blues')
    plt.xticks(range(len(labels_order)), labels_order)
    plt.yticks(range(len(labels_order)), labels_order)
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix")

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, str(cm[i, j]), ha='center', va='center', color='black', fontsize=12)

    plt.show()


# -----------------------------
# Ejecutar
# -----------------------------
if __name__ == "__main__":
    args = Args()
    main(args)
```
